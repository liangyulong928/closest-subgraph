```python
import spartan as st
```

```python
tensor_data = st.loadTensor(path = "data/email-EU.csv",sep=",",header=None)
tensor_data
```

```python
stensor = tensor_data.toSTensor(hasvalue=True)
```

```python
stensor._data
```

```python
graph = st.Graph(stensor, bipartite=True, weighted=True, modet=2)
```

```python
hs = st.HoloScope(graph)
```

```python
print(hs)
```

```python
res = hs.run(level=0, k=1)
```

```python
print(res[0][1])
```

```python
print(f"可以区块横坐标区间:[{min(res[0][0][0])},{max(res[0][0][0])}]")
print(f"可以区块纵坐标区间:[{min(res[0][0][1])},{max(res[0][0][1])}]")
print(f"密度为:{res[0][1]}")
```
