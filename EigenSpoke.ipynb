```python
import spartan as st
```

```python
tensor_data = st.loadTensor(path = "data/email-EU.csv",sep=",",header=None)
stensor = tensor_data.toSTensor(hasvalue=True)
stensor._data
```

```python
es = st.Eigenspokes(stensor)
```

```python
res = es.run()
```

```python
print(f"检测异常区间为：[{min(res)},{max(res)}]")
print(f"异常区间密度为：{(max(res)-min(res))/len(res)}")
```
